```python
import time
from selenium import webdriver
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
phone_brands=['m-iPhone', 'm-Infinix', 'm-realme', 'm-Xiaomi', 'm-OPPO', 'm-Samsung']
```

```python
empty_dataframe1 = pd.DataFrame()

for brand in range(0, len(phone_brands)):
    for page in range(1,6):
        
        root_url=f"view-source:https://www.bukalapak.com/c/handphone/hp-smartphone/{phone_brands[brand]}?page={page}"
        driver = webdriver.Chrome(executable_path='C:/Users/User/Downloads/chromedriver_win32/chromedriver.exe')
        driver.get(root_url)
        root_read = driver.page_source
        
        
        awal=root_read.find('{"products"')
        akhir=root_read.find(',"facets":')
        product_result=root_read[awal:akhir]
        product_result=product_result + '}'
        
        data_root_json=json.loads(product_result)
        
        for i in range (0, len(data_root_json['products'])):
            items_data=data_root_json['products'][i]
            items_data=json_normalize(items_data)
            df=pd.DataFrame(items_data)
            
            df['brand']=phone_brands[brand][2:]
            df['rootcategory_name']=items_data['category.structure'][0][1]
            df=df[['id', 'sku_id', 'name', 'price', 'discount_percentage', 'rootcategory_name', 'brand', 'rating.average_rate', 'rating.user_count',  'store.address.province']]
            empty_dataframe1=empty_dataframe1.append(df, ignore_index = True)    



#         print(phone_brands[brand], page)
```

```python

def execute_mogrify(conn, df, table):
    """
    Using cursor.mogrify() to build the bulk insert query
    then cursor.execute() to execute the query
    """
    # Create a list of tupples from the dataframe values
    tuples = [tuple(x) for x in df.to_numpy()]
    
#     print(tuples)
    # Comma-separated dataframe columns
    cols = ','.join(list(df.columns))
    
    
#     print(cols)
    # SQL quert to execute
    cursor = conn.cursor()
    values = [cursor.mogrify("(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)", tup).decode('utf8') for tup in tuples]
    query  = "INSERT INTO %s(%s) VALUES " % (table, cols) + ",".join(values)
    
    try:
        cursor.execute(query, tuples)
        conn.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        conn.rollback()
        cursor.close()
        return 1
    print("execute_mogrify() done")
    cursor.close()
```

```python
# Connect to the database
conn = connect(conn_params_dic)

# Run the execute_many strategy
execute_mogrify(conn, df, 'please')
```
